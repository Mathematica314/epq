##### Initialisation

```python
import os
import torchaudio
import torch
import kernmanager
from collections import defaultdict
import pickle

data = os.fsencode("data")
kerns = os.path.join(data, os.fsencode("kerns"))
midi = os.path.join(data,os.fsencode("midi"))
wavs = os.path.join(data,os.fsencode("wavs"))
features = os.path.join(data,os.fsencode("features"))
spectrograms = os.path.join(data,os.fsencode("spectrograms"))
```

##### Note frequencies

```python
notefreqdata = defaultdict(list)
for composer in os.listdir(kerns):
    for file in os.listdir(os.path.join(kerns, composer)):
        notefreqdata[composer.decode()].append((file.decode(), (kernmanager.notefreq(os.path.join(kerns, composer, file).decode()))))

nfs = {k:{x[0]:dict(x[1]) for x in notefreqdata[k]} for k in notefreqdata.keys()}
with open("notefreqs.pkl","wb") as file:
    pickle.dump(nfs, file)
```

##### Midi

```python
for composer in os.listdir(kerns):
    for file in os.listdir(os.path.join(kerns,composer)):
        print(file)
        kernmanager.midi(os.path.join(kerns, composer, file).decode(), os.path.splitext(os.path.join(midi, composer, file).decode())[0] + ".mid")
```

##### Waveforms

```python
import midi2audio

fs = midi2audio.FluidSynth("[GD] Steinway Model D274.sf2")
for composer in os.listdir(midi):
    for file in os.listdir(os.path.join(midi,composer)):
        fs.midi_to_audio(os.path.join(midi, composer, file).decode(), os.path.splitext(os.path.join(wavs, composer, file).decode())[0] + ".wav")
```

##### Features

```python
for composer in os.listdir(kerns):
    for file in os.listdir(os.path.join(kerns,composer)):
        fes = kernmanager.get_features(os.path.join(kerns, composer, file).decode())
        with open(os.path.splitext(os.path.join(features,composer,file).decode())[0]+".pkl","wb") as file:
            pickle.dump(fes,file)
feature_data = []
for composer in os.listdir(features):
    for file in os.listdir(os.path.join(features, composer)):
        with open(os.path.join(features, composer, file).decode(), "rb") as f:
            comp = pickle.load(f)
            feature_data.append((composer.decode(), comp))
with open("fets.pkl", "wb") as file:
    pickle.dump(feature_data, file)
```

##### Spectrograms

```python
for composer in os.listdir(wavs):
    for file in os.listdir(os.path.join(wavs, composer)):
        print(file)
        waveform,sample_rate = torchaudio.load(os.path.join(wavs, composer, file).decode(), normalize=True)
        waveform = torch.mean(waveform,dim=0,keepdim=True)
        spectrogram = torchaudio.transforms.MelSpectrogram(sample_rate*0.02,n_mels=256)(waveform)
        with open(os.path.splitext(os.path.join(spectrograms,composer,file).decode())[0]+".pkl","wb") as file:
            pickle.dump(spectrogram[0],file)
```
